```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("Semanas", color="purple")
plt.ylabel("Ahorro($)", color="purple")
plt.title("Ahorro acumulado por semana", color="purple")
semanas = [0, 1, 2, 3, 4, 5]
ahorro = [50 * s for s in semanas]  
plt.plot(semanas, ahorro, "mo--")
```

El gráfico demuestra el ahorro de una persona que guarda 50$ por semana. Es representado por un gráfico lineal ya que muestra de manera clara el crecimeinto de los ahorros. 

```python
plt.xlabel("Tiempo(hora)", fontweight="bold", color="pink")
plt.ylabel("Notas", fontweight="bold", color="pink")
plt.title("Cantidad de horas estudiadas y notas obtenidas", fontweight="bold", color="pink")
horas = [1, 2, 3, 4, 5, 6]
notas = [40, 50, 55, 65, 78, 85]
plt.scatter(horas, notas, c="pink", marker="D")
```

Representa la relación entre las horas de estudio y las calificaciones obtenidas por los alumnos. Se eligió este gráfico para observar si existe una correlación entre ambas variables.

```python
plt.xlabel("Mes", color="blue")
plt.ylabel("Cantidad de libros leídos", color="blue")
plt.title("Lectura mensual", color="blue")
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo"]
libros = [3, 4, 2, 5, 6]
plt.bar(meses, libros, color ="skyblue")
```

Muestra la cantidad de libros leídos en cada mes del año(desde Enero, hasta Mayo). Se eligió esta visualización porque permite comparar fácilmente los valores entre los distintos meses.

```python
plt.xlabel("Horas por día", color="magenta")
plt.ylabel("Cantidad de personas", color="magenta")
plt.title("Uso diario de redes sociales", color="magenta")
uso_redes = [1, 2, 2.5, 3, 3.5, 1.5, 2, 4, 5, 3, 2.5, 1, 4.5, 3, 2, 2.5, 3.5, 4, 5, 1.5]
plt.hist(uso_redes, bins=5,edgecolor="black", color="pink")
```

Este gráfico representa la distribución de tiempo que las personas pasan en redes sociales cada día. Se usó un histograma para analizar la frecuencia de diferentes rangos de tiempo.

```python
plt.title("Comida favorita de 5 amigas")
comidas = ["Tarta de verduras", "Pasta", "Sushi"]
cantidades = [1, 3, 1]
#Lista de colores(para cada categoría)
colores = ["lightgreen","lightcoral", "lightblue"]
plt.pie(cantidades, labels=comidas, colors=colores)
```

Muestra la proporción de 5 amigas que prefieren distintos tipos de comida. Se eligió esta visualización para resaltar cómo se divide el total entre varias categorías.
